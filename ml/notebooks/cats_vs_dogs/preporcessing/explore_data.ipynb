Load and explore the data

Import libraries

```python
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import cv2
```

Load the data

```python
#load data and metadata of cats_vs_dogs from tensorflow datasets 
data, metadata = tfds.load('cats_vs_dogs', as_supervised=True, with_info=True)
```

Show images from the dataset

```python
#Show 5 images of simple form
tfds.as_dataframe(data['train'].take(5), metadata)
```

```python
#Show resize and tranformed images
IMG_SIZE = 100

plt.figure(figsize=(20,20))
for i, (image, label) in enumerate(data['train'].take(20)):
    image = cv2.resize(image.numpy(), (IMG_SIZE, IMG_SIZE))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    plt.subplot(4, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(image, cmap='gray')
```
